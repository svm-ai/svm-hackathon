```python
import zipfile
import numpy as np
import pandas as pd
import os
import torch
```

```python
with zipfile.ZipFile('msa_embeddings_full_60K.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
seq_acc = pd.read_csv('/home/ubuntu/proteinbind/msa_accession.csv', header=None)
seq_acc.rename(columns={0: 'filename', 1: 'sequence'}, inplace=True)
seq_acc
```

```python
dir_path = 'msa_embeddings/'
data = []

for filename in os.listdir(dir_path):
    try:
        # Load the numpy array
        array = torch.load(os.path.join(dir_path, filename))

        # Add the filename and array data to the list
        row = [filename] + array.tolist()
        data.append(row)
    except EOFError:
        print(f"Error loading file {filename}: EOFError (Ran out of input)")
    except Exception as e:
        print(f"Error loading file {filename}: {str(e)}")
        continue
```

```python
# Create a DataFrame from the list
df = pd.DataFrame(data)

# Set the column names
df.columns = ['filename'] + ['col_{}'.format(i) for i in range(1, df.shape[1])]
```

```python
# Print the DataFrame
df
```

```python
df['filename'] = df['filename'].str.replace('.t', '')
```

```python
merged_df = pd.merge(seq_acc, df, on='filename')
```
