```python
import zipfile
import numpy as np
import pandas as pd
import os
import torch
```

```python
# Open embedded GO Terms
with zipfile.ZipFile('go_embeddings_20k.zip', 'r') as zip_ref:
    zip_ref.extractall()
```

```python
# Compile embedded GO Terms into Pandas Dataframe
dir_path = 'go_embeddings/'
# Initialize an empty list to store the data
data = []

# Iterate over all .npy files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.npy'):
        # Load the numpy array
        array = np.load(os.path.join(dir_path, filename))

        # Add the filename and array data to the list
        row = [filename] + array.tolist()
        data.append(row)

# Create a DataFrame from the list
GO = pd.DataFrame(data)

# Set the column names
GO.columns = ['filename'] + ['col_{}'.format(i) for i in range(1, GO.shape[1])]

# Print the DataFrame
GO    # Load the numpy array
```

```python
GO['filename'] = GO['filename'].str.replace('.npy', '')
```

```python
dir_path = 'GO_AA_Embeddings/'
# Initialize an empty list to store the data
data = []

# Iterate over all .npy files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.pt'):
        # Load the numpy array
        array = torch.load(os.path.join(dir_path, filename))

        # Add the filename and array data to the list
        row = [filename] + array.tolist()
        data.append(row)

# Create a DataFrame from the list
AA = pd.DataFrame(data)

# Set the column names
AA.columns = ['filename'] + ['col_{}'.format(i) for i in range(1, AA.shape[1])]

# Print the DataFrame
AA    # Load the numpy array
```

```python
AA['filename'] = AA['filename'].str.replace('.pt', '')
```
